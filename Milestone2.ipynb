DATA PREPROCESSING AND MODEL TRAINING.

We have scaled the attributes required for the model using a Standard scaler and have encoded the categorical attributes using a Label encoder before training the model.
Also we have selected the attributes Age,Condiiton,Drug,Ease of use,Effectiveness,Satisfaction score,Sex,Useful count to predict the target Side Effects.
We have also split the dataset into training and test sets in ratio 17:3 for effective prediction and to improve accuracy.
The classification Model used hereto fit is a Decision Tree classifier which uses Entropy and Gain to classify the Attributes.

```python
X=drugs.iloc[:,[0,1,3,5,6,8,9,11]].values
y=drugs.Sides
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
le=LabelEncoder()
X[:,0]=le.fit_transform(X[:,0])
X[:,1]=le.fit_transform(X[:,1])
X[:,2]=le.fit_transform(X[:,2])
X[:,6]=le.fit_transform(X[:,6])
y=le.fit_transform(y)
x_train, x_test , y_train , y_test= train_test_split(X,y,test_size=0.15,random_state=1)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf=clf.fit(x_train,y_train)

```
